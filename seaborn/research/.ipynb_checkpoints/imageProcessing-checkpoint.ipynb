```python
# plotting multiple images.
fig, axs = plt.subplots(nrows=2, ncols=2 ,figsize=(10,10))
axs[0][0].set_title('graycurls')
axs[0][0].imshow(graycurls)


axs[0][1].set_title('graycurlsinvert')
axs[0][1].imshow(graycurlsinvert)

axs[1][0].set_title('curls')
axs[1][0].imshow(curls)

axs[1][1].set_title('invert')
axs[1][1].imshow(invert)
plt.show()
```

```python
#applying hysteresis thresholding.
image = np.array([1, 2, 3, 2, 1, 2, 1, 3, 2])
hyster_curls=ski.filters.apply_hysteresis_threshold(curls, 100, 250).astype(int)
```

```python
plt.imshow(hyster_curls)
```

```python
## butterworth filter
# apply a high pass and low pass filter to grayscale and color image
```

```python
high_pass=butterworth(curls,0.07,True,8)
low_pass=butterworth(curls,0.04,False,4,channel_axis=-1)
```

```python
# plotting multiple images.
fig, axs = plt.subplots(nrows=1, ncols=3 ,figsize=(10,10))
axs[0].set_title('curls')
axs[0].imshow(curls)


axs[1].set_title('high_pass')
axs[1].imshow(high_pass)

axs[2].set_title('low_pass')
axs[2].imshow(low_pass)


plt.show()
```

```python
### sobel filter, helps detect edges
sobel=filters.sobel(curls)
```

```python
# plotting multiple images.
fig, axs = plt.subplots(nrows=1, ncols=2 ,figsize=(10,10))
axs[0].set_title('curls')
axs[0].imshow(curls)


axs[1].set_title('sobel')

axs[1].imshow(sobel)
plt.show()
```

```python
boxcar=window('boxcar',(100,100))
triang=window('triang',(100,100))
blackman=window('blackman',(100,100))
hamming=window('hamming',(100,100))
parzen=window('parzen',(100,100))
nuttall=window('nuttall',(100,100))
```

```python
# plotting multiple images.
fig, axs = plt.subplots(nrows=1, ncols=6 ,figsize=(15,15))
axs[0].set_title('boxcar')
axs[0].imshow(boxcar)

axs[1].set_title('triang')
axs[1].imshow(triang)

axs[2].set_title('blackman')
axs[2].imshow(blackman)

axs[3].set_title('hamming')
axs[3].imshow(hamming)

axs[4].set_title('parzen')
axs[4].imshow(parzen)

axs[5].set_title('nuttall')
axs[5].imshow(nuttall)

plt.show()
```
