```python
# curl image 
curls=ski.io.imread("data/data/curly/00cbad1ffe22d900018e5a2e7376daed4.jpg")
```

```python
type(curls)
```

```python
plt.imshow(curls)
```

```python
# conversion from rgb to gray
```

```python
graycurls=ski.color.rgb2gray(curls)
```

```python
plt.imshow(graycurls)
```

```python
#invert image
invert=ski.util.invert(curls)
plt.imshow(invert)
```

```python
graycurlsinvert=ski.util.invert(graycurls)
plt.imshow(graycurlsinvert)
```

```python
# plotting multiple images.
fig, axs = plt.subplots(nrows=2, ncols=2 ,figsize=(10,10))
axs[0][0].set_title('graycurls')
axs[0][0].imshow(graycurls)


axs[0][1].set_title('graycurlsinvert')
axs[0][1].imshow(graycurlsinvert)

axs[1][0].set_title('curls')
axs[1][0].imshow(curls)

axs[1][1].set_title('invert')
axs[1][1].imshow(invert)
plt.show()
```
