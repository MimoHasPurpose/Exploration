```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,5,100)
```

```python
t
```

```python
x=np.sin(t)
```

```python
print(x)
```

```python
# resample a signal to n points using fft
x_resample=sp.signal.resample(x,25)
```

```python
plt.plot(t,x)
plt.plot(t[::4],x_resample,'ko')
```
