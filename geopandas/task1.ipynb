#### A GeoSeries is essentially a vector where each entry in the vector is a set of shapes corresponding to one observation.
#### An entry may consist of only one shape (like a single polygon) or multiple shapes that are meant 
#### to be thought of as one observation (like the many polygons that make up the State of Hawaii or a country like Indonesia).

### geopanas has 3 basic classes of geometric objects:
- points
- lines
- Polygons
